```python
!pip install requests_html
#from google.colab import drive
#drive.mount("/drive")
#from google.colab import files
#uploaded = files.upload()
import io
#Google Colab Request
import requests  #Thực thi HTTP Request
import datetime  #Làm việc với ngày tháng năm
import pandas as pd #API phân tích và xử lý dữ liệu
import numpy as np 
import array as arr
#Thư viện lấy dữ liệu từ web
from requests_html import HTML 
from requests_html import HTMLSession
#Thư viện trích xuất dữ liệu từ file HTML và XML
from bs4 import BeautifulSoup
```

```python
#Lấy dữ liệu từ chợ tốt 
#Thu thập dữ liệu)
ar = []
for page in range(1000):
    page= str(page+1)
    link ="https://nha.chotot.com/tp-ho-chi-minh/mua-ban-can-ho-chung-cu?page="+page
    r = requests.Session()
    rl = r.get(link)
    # r chứa thông tin về session của từng đường link
    soup = BeautifulSoup(rl.text, 'lxml')
    # Soup sử dụng parser 'lxml'
    a = soup.find_all("a",class_='adItem___2GCVQ')
    # Tìm các thẻ <a> - thẻ hyperlink với class là 'adItem___2GCVQ'. Khu vực này chứa link cụ thể của các bất động sản
    for i in range(len(a)):
        link ="https://nha.chotot.com/"+a[i]['href']
        ar.append(link)
        print(ar[i])
    # Mảng link sẽ chứa link của tất cả bất động sản
print("done")
df = pd.DataFrame(ar,columns=['links'])


#df.style.set_properties(subset=['links'], **{'width': '600px'})
#Tạo một DataFrame với tất cả các link thu được
```

```python
# Xuất file csv
df.to_csv('Link_ChoTot.csv',index=True)
```
